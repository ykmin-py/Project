```python
# clean the tweet function
ps = nltk.PorterStemmer()
wn = nltk.WordNetLemmatizer()

def cleanTweet(text):
    text = re.sub(r'@[A-Za-z0-9]+', " ", text)
    text = re.sub(r'RT[\s]+', " ", text)
    text = re.sub(r'https?:\/\/\S+', " ", text)
    text  = "".join([char.lower() for char in text if char not in string.punctuation])
    text = re.split('\W+', text)
    #text = [ps.stem(word) for word in text]   
    text = [wn.lemmatize(word) for word in text]   
    text = (" ".join(text)).strip()
    text_final = text

    return text_final
```

```python
# cleaning the tweet
df['Tweets'] = df['Tweets'].apply(cleanTweet)
```

# Data Analysis 

```python
# get the subjectivity, polarity, sediment analysis
def GetSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def GetPolarity(text):
    return TextBlob(text).sentiment.polarity

def GetPositiveAnalysis(polarity):
    if polarity > 0.0:
        return "Positive"
    if polarity == 0.0:
        return "Neutral"
    if polarity < 0.0:
        return "Negative"

def GetSedimentLabel(polarity):
    if polarity > 0.0:
        return 1
    if polarity == 0.0:
        return 0
    if polarity < 0.0:
        return -1
# adding labels, subjectivity, polarities
df['Subjectivity'] = df['Tweets'].apply(GetSubjectivity)
df['Polarity'] = df['Tweets'].apply(GetPolarity)
df['Analysis'] = df['Polarity'].apply(GetPositiveAnalysis)
df['Label'] = df['Polarity'].apply(GetSedimentLabeldimentLabel)

df.head()
```

```python
# function to calculate percentage
def percentage(part, whole):
        temp = 100 * float(part) / float(whole)
        return format(temp, '.2f')
```

```python
# calculate the percentage of positive, neutral, negative percentages of tweets
positive = 0
polarity = 0
neutral = 0
negative = 0

for tweet in df['Tweets']:
    #print(tweet.t)
    analysis = TextBlob(tweet)
    polarity += analysis.sentiment.polarity
    
    if (analysis.sentiment.polarity > 0.00):
        positive +=1
        
    if (analysis.sentiment.polarity < 0.00):
        negative +=1
        
    if (analysis.sentiment.polarity == 0.00):
        neutral +=1

TotalSearchNum = df.shape[0]
positive = percentage(positive, TotalSearchNum)
negative = percentage(negative, TotalSearchNum)
neutral = percentage(neutral, TotalSearchNum)
```

# part 1: Apply Support Vector Classifier for Sediment Analysis

```python
# test and train split

from sklearn.model_selection import train_test_split

y = df['Label']

x_train, x_test, y_train, y_test = train_test_split(df['Tweets'], y, stratify=y, random_state=1, test_size=0.3, shuffle=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# vectorize tweets using CountVector & removing stopwords

vectorizer = CountVectorizer(binary=True, min_df = 2, max_df = 40)
vectorizer.fit(list(x_train) + list(x_test))

x_train_vec = vectorizer.transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```

```python
# visualize removed stop words
vectorizer.stop_words_
```

```python
# Apply Support Vector Classifier to train model and make predictions

from sklearn import svm

svm = svm.SVC(kernel = 'linear' , probability = True)
# fit the SVC model based on the training data

prob = svm.fit(x_train_vec,y_train).predict_proba(x_test_vec)

y_pred_svm = svm.predict(x_test_vec)
```

```python
# Accuracy Score for SVC

from sklearn.metrics import accuracy_score
print("Accuracy score for SVC is:", accuracy_score(y_test, y_pred_svm)*100, '%')
```

# Part 2: Apply Unsupervised ML for Clustering Analysis

```python
# import kmeans, 3 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, n_jobs = -1, max_iter =50000, random_state = 1)

# vectorize tweet, and utilized as input for clustering analysis
Vector_data = vectorizer.transform(df['Tweets'])
kmeans.fit(Vector_data)
```

```python
# shows the results of clustering
y_km = kmeans.fit_predict(Vector_data)
```

```python
y_km
```

```python
# print percentages of each cluster

cluster_1 = 0
cluster_2 = 0
cluster_3 = 0

for cluster in y_km:
    
    if (cluster == 0.00):
        cluster_1 +=1
        
    if (cluster == 1):
        cluster_2 +=1
        
    if (cluster == 2):
        cluster_3 +=1

TotalSearchNum = len(y_km)
cluster_1 = percentage(cluster_1, TotalSearchNum)
cluster_2 = percentage(cluster_2, TotalSearchNum)
cluster_3 = percentage(cluster_3, TotalSearchNum)
```

```python
# define pie chart to plot percentages of cluster
def plotPieChart_cluster(positive, neutral, negative, SearchKeyWrods, TotalSearchNum):
        labels = ['Cluster_1 [' + str(positive) + '%]', 'Cluster_2 [' + str(neutral) + '%]',
                  'Cluster_3 [' + str(negative) + '%]']
        sizes = [positive, neutral, negative]
        colors = ['green', 'gold', 'red']
        patches, texts = plt.pie(sizes, colors=colors, startangle=90)
        plt.legend(patches, labels, loc="best")
        plt.title("people's reacting on covid vaccine"  + "by analyzing "  + str(TotalSearchNum) + " Tweets." + "using clustering analysis")
        plt.axis('equal')
        plt.tight_layout()
        plt.show()

plotPieChart_cluster(cluster_1,cluster_2,cluster_3,SearchKeyWrods, TotalSearchNum)
```

```python
# print out feature words in each cluster
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```
